# PlotCutResults
This notebook it intended to selelect the signal region cuts we will use in the B-L stop search.
The input to this notebook is a directory of text files from our TestOptimize script.
## Setup
First, we include some python moduldes to set up our environment

```python
%matplotlib inline
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
import re
import itertools

mp.rcParams['text.usetex']=True
```

## Input files
The input to this notebook is a directory of text files from our TestOptimize script.
Each file in this directory corresponds to a set of cuts and choice of signal model, and includes a line for each of
- Number of signal events
- Number of background events
- Significance estimated using $\frac{s}{\sqrt{s+b}}$
- Significance estimated using $Z_N$

## Define cuts and useful lists
In this block, we first define the list of cuts we will be testing.
Then, we construct the selection tags by taking all the combinations of $h_T$, $m_{b\ell}$ asymmetry, and $E_{T}^{\mathrm{miss}}$ significance cuts.
Finally, we constuct a mass axis, and a placeholder numpy array for num_sig, num_bkg, and significance for each selection combination.

```python
# first define the cut values which we will use
ht_cuts       = [1000, 1100, 1200]
mbl_asym_cuts = [0.3, 0.35, 0.4]
# mbl_asym_cuts = [0.4]
met_sig_cuts  = [10, 15, 20, 100]
# met_sig_cuts  = [15, 10]

# constuct the selection tags
selections = []
for hc, mac, msc in itertools.product(ht_cuts, mbl_asym_cuts, met_sig_cuts):
    selections.append('ht_%s__mbl_asym_%s__met_sig_%s' % (hc, mac, msc))

# create axes for plotting!
# one mass axis
mass_axis = np.linspace(100, 1000, 10)

# one np array for each variable per each selection
num_sig_arrays = {sel:np.zeros(len(mass_axis)) for sel in selections}
num_bkg_arrays = {sel:np.zeros(len(mass_axis)) for sel in selections}
soverb_arrays  = {sel:np.zeros(len(mass_axis)) for sel in selections}
zn_arrays      = {sel:np.zeros(len(mass_axis)) for sel in selections}
```

## readCutResultFile
This function is used to read a file, and extract its information.

```python
def readCutResultFile(file_name):
    this_file = file(file_name)
    
    num_bkg = 0
    num_sig = 0
    soverb  = 0
    zn      = 0
    for i, l in enumerate(this_file.readlines()):
        splits = l.rstrip().split()
        
        if   i == 0: num_bkg = splits[1]
        elif i == 1: num_sig = splits[1]
        elif i == 2: soverb  = splits[1]
        elif i == 3: zn      = splits[1]
    this_file.close()
    
    return {'num_bkg':num_bkg, 'num_sig':num_sig, 'soverb':soverb, 'zn':zn}
```

## Read all the files in our directory, and fill our y arrays

```python
for sel in selections:
    for i, mass_val in enumerate(mass_axis):
        file_name = 'results/cut_result_info__ttbar_T%d__%s.txt' % (mass_val, sel)
        results = readCutResultFile(file_name)
        
        num_sig_arrays[sel][i] = results['num_sig']
        num_bkg_arrays[sel][i] = results['num_bkg']
        soverb_arrays[sel][i]  = results['soverb']
        zn_arrays[sel][i]      = results['zn']
```

# Plotting
## getLabel
First, we define a function to take a selection tag, and return the label which is used in the legend.

```python
def getLabel(sel):
    # return sel.replace('ht_', '$h_{T}$:').replace('mbl_asym_', 'mbl asym:').replace('met_sig_', 'met sig:').replace('__', ' - ')
    return sel.replace('ht_', '').replace('mbl_asym_', '').replace('met_sig_', '').replace('__', ' - ')
```

## pilePlots
This is a helper function to draw a list of plots on the same axes.
It takes as input:
- x_points: Numpy array of x points (there is only one list of x points for the figure)
- y_points: Dictionary of numpy arrays of y points (one for each selection level)
- selections_to_plot: List of selection levels to plot (these should correspond to the keys in the y_points dictionary)

```python
def pilePlots(x_points, y_points, selections_to_plot=selections):
    for stp in selections_to_plot:
        this_label = getLabel(stp)
        plt.plot(x_points, y_points[stp], label=this_label)
```

## drawTwoPilesSideBySideAndLegend
This function makes two plots side-by-side, and then adds a legend to the right side of the figure.
The function takes as input:
- x_points: Numpy array of x points (there is only one list of x points for the figure)
- y_points_left: Dictionary of numpy arrays representing the y points on the left plot
- y_points_right: Dictionary of numpy arrays representing the y points on the right plot
- x_label: X-axis label
- y_label_left: Left y-axis label
- y_label_right: Right y-axis label
- selections_to_plot: List of selection levels to plot (these should correspond to the keys in the y_points dictionary)
- fig_size: Size of the figure in inches provided as a tuple
- log_left: Is the plot on the left in log scale
- log_right: Is the plot on the right in log scale
- leg_title: Title of the legend
- leg_ncol: number of columns in the legend

```python
def drawTwoPilesSideBySideAndLegend( x_points
                                   , y_points_left
                                   , y_points_right
                                   , x_label
                                   , y_label_left
                                   , y_label_right
                                   , selections_to_plot=selections
                                   , fig_size=(20,5)
                                   , log_left=False
                                   , log_right=False
                                   , leg_title=None
                                   , leg_ncol=1
                                   ):
    fig = plt.figure(figsize=fig_size)
    
    # left plot
    plt.subplot(1,3,1)
    pilePlots(x_points, y_points_left, selections_to_plot=selections_to_plot)
    plt.xlabel(x_label)
    plt.ylabel(y_label_left)
    if log_left:
        plt.yscale('log')
        
    # right plot
    plt.subplot(1,3,2)
    pilePlots(x_points, y_points_right, selections_to_plot=selections_to_plot)
    plt.xlabel(x_label)
    plt.ylabel(y_label_right)
    if log_right:
        plt.yscale('log')

    # add legend
    plt.legend( bbox_to_anchor=(1.05, 1)
              , loc=2
              , borderaxespad=0
              , ncol=leg_ncol
              , title=leg_title
              )
```

## drawFourPilesInGridAndLegend
This function makes four plots in a 2x2 grid, and then adds a legend to the right side of the figure.
The function takes as input:
- x_points: Numpy array of x points (there is only one list of x points for the figure)
- y_points_ul: Dictionary of numpy arrays representing the y points on the upper left plot
- y_points_ur: Dictionary of numpy arrays representing the y points on the upper right plot
- y_points_ul: Dictionary of numpy arrays representing the y points on the lower left plot
- y_points_ur: Dictionary of numpy arrays representing the y points on the lower right plot
- x_label: X-axis label
- y_label_ul: Upper left y-axis label
- y_label_ur: Upper right y-axis label
- y_label_ll: Lower left y-axis label
- y_label_lr: Lower right y-axis label
- selections_to_plot: List of selection levels to plot (these should correspond to the keys in the y_points dictionary)
- fig_size: Size of the figure in inches provided as a tuple
- log_ul: Is the plot on the upper left in log scale
- log_ur: Is the plot on the upper right in log scale
- log_ll: Is the plot on the lower left in log scale
- log_lr: Is the plot on the lower right in log scale
- leg_title: Title of the legend
- leg_ncol: number of columns in the legend
- zoom_ul: Zoom the upper left axes
- zoom_ur: Zoom the upper right axes
- zoom_ll: Zoom the lower left axes
- zoom_lr: Zoom the lower right axes

```python
def drawFourPilesInGridAndLegend( x_points
                                , y_points_ul
                                , y_points_ur
                                , y_points_ll
                                , y_points_lr
                                , x_label
                                , y_label_ul
                                , y_label_ur
                                , y_label_ll
                                , y_label_lr
                                , selections_to_plot=selections
                                , fig_size=(20,10)
                                , log_ul=False
                                , log_ur=False
                                , log_ll=False
                                , log_lr=False
                                , leg_title=None
                                , leg_ncol=1
                                , zoom_ul=None
                                , zoom_ur=None
                                , zoom_ll=None
                                , zoom_lr=None
                                ):
    fig = plt.figure(figsize=fig_size)

    # Upper left
    plt.subplot(2, 3, 1)
    pilePlots(x_points, y_points_ul, selections_to_plot=selections_to_plot)
    plt.xlabel(x_label)
    plt.ylabel(y_label_ul)
    if log_ul:
        plt.yscale('log')
    if zoom_ul:
        plt.axis(zoom_ul)

    # Upper right
    plt.subplot(2, 3, 2)
    pilePlots(x_points, y_points_ur, selections_to_plot=selections_to_plot)
    plt.xlabel(x_label)
    plt.ylabel(y_label_ur)
    if log_ur:
        plt.yscale('log')
    if zoom_ur:
        plt.axis(zoom_ur)

    # add legend
    plt.legend( bbox_to_anchor=(1.05, 1)
              , loc=2
              , borderaxespad=0
              , ncol=leg_ncol
              , title=leg_title
              )

    # Lower left
    plt.subplot(2, 3, 4)
    pilePlots(x_points, y_points_ll, selections_to_plot=selections_to_plot)
    plt.xlabel(x_label)
    plt.ylabel(y_label_ll)
    if log_ll:
        plt.yscale('log')
    if zoom_ll:
        plt.axis(zoom_ll)

    # Lower right
    plt.subplot(2, 3, 5)
    pilePlots(x_points, y_points_lr, selections_to_plot=selections_to_plot)
    plt.xlabel(x_label)
    plt.ylabel(y_label_lr)
    if log_lr:
        plt.yscale('log')
    if zoom_lr:
        plt.axis(zoom_lr)
```
